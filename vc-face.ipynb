# Regiao de Interesse

Regiao de interesse (ROI)  a rea onde queremos trabalhar em nosso algoritmo para realizar algum processamento de viso computacionam, pode ocorrer em uma ou mais partes especficas da imagem.

```python
pessoas = cv2.imread("imagens/px-people.jpg")
pessoas = cv2.cvtColor(pessoas, cv2.COLOR_BGR2RGB)
plt.imshow(pessoas)
```

Extraindo um ROI

```python
img_roi = pessoas[100:210,1090:1200]
plt.imshow(img_roi)
```

```python
# necessrio a conversao a bgr para nao ter problemas na cor da imagem
img_roi_bgr = cv2.cvtColor(img_roi, cv2.COLOR_RGB2BGR) 
cv2.imwrite("./imagens/img_roi.png",img_roi_bgr)
```
